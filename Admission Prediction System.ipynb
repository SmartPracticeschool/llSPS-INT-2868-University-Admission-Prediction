```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
#First we will see the accuracy using linear regression algorithm 

from sklearn.linear_model import LinearRegression
lr=LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
y_predict=lr.predict(x_test)
y_predict 
```

```python
lr.predict([[13,326,117,4,3.5,4,6.9,0]])
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_predict)
```

```python
#now we will see the accuracy using random forest algorithm
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor(n_estimators=10 ,random_state=0, n_jobs=-1)  #n_estimators tells no of dicision trees
rf.fit(x_train,y_train)
```

```python
y_pred = rf.predict(x_test)
y_pred
```

```python
y_test
```

```python
rf.predict([[13,326,117,4,3.5,4,6.9,0]])
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```

```python
#now will see the accuracy using decision tree 

from sklearn.tree import DecisionTreeRegressor
dt = DecisionTreeRegressor(random_state=0,criterion="mae")
dt.fit(x_train,y_train)
```

```python
y_preddd = dt.predict(x_test)
y_preddd
```

```python
dt.predict([[13,326,117,4,3.5,4,6.9,0]])
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_preddd)
```

```python
Methods = ['Linear Regression', 'Decision Trees', 'Random Forests']
Scores = np.array([r2_score(y_test,y_predict), r2_score(y_test,y_preddd), r2_score(y_test,y_pred)])

fig, ax = plt.subplots(figsize=(8,6))
sns.barplot(Methods, Scores)
plt.title('Algorithm Prediction Accuracies')
plt.ylabel('Accuracy')
```
